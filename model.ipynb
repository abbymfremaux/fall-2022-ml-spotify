```python
import pandas as pd
tracks_df = pd.read_csv('SpotGenTrack/Data Sources/spotify_tracks.csv')

tracks_df.head(10)
```

```python
# see all columns in the dataframe
tracks_df.columns
```

```python

from sklearn.cluster import KMeans
model = KMeans(n_clusters = 10, init = 'k-means++', random_state = 42)
model.fit(tracks_df[['energy', 'valence', 'danceability', 'instrumentalness', 'tempo', 'acousticness', 'loudness', 'speechiness', 'liveness', 'key']])

print(model.labels_[:10], '\n\n', min(model.labels_), max(model.labels_))

tracks_df['type'] = model.labels_
tracks_df.head(10)

tracks_df.to_csv('result.csv', index=False)
```
